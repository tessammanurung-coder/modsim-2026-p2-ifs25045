## Distribusi jawaban kusioner dari keseluruhan data.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("data_kuesioner.xlsx")
data = df.iloc[:, 1:]

overall = data.stack().value_counts()

plt.figure()
overall.plot(kind='bar')
plt.title("Distribusi Jawaban Kuesioner (Keseluruhan)")
plt.xlabel("Kategori Jawaban")
plt.ylabel("Jumlah")
plt.show()
```

## Proporsi jawaban kusioner dari keseluruhan data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("data_kuesioner.xlsx")
data = df.iloc[:, 1:]
all_answers = data.stack().dropna()
skala = ["SS", "S", "CS", "CTS", "TS", "STS"]

all_answers = all_answers[all_answers.isin(skala)]
counts = all_answers.value_counts()
proporsi = counts / counts.sum() * 100
proporsi = proporsi.reindex(skala).dropna()

plt.figure(figsize=(7,7))
plt.pie(
    proporsi,
    labels=proporsi.index,
    autopct="%.1f%%",
    startangle=90,
    wedgeprops={"edgecolor": "white"}
)
plt.title("Proporsi Jawaban Kuesioner dari keseluruhan data (SS  STS)")
plt.axis("equal")
plt.show()
```
